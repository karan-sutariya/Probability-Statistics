```python
'''
Name: Karan Sutariya
Roll No: 2103322
CS230 Assignment[4]: Part II - Q2 Simulating Poisson process and plotting a histogram of the time of k-th arrival

'''


#Importing the following libraries to work with poisson processes and plotting the distribution


import numpy as np
from numpy.random import exponential
import matplotlib.pyplot as plot

rate = 2#per hour

alpha = 1/rate

k = 7 #fixating the value of k

sample = np.zeros(100000) #creating an (100000 x 1) zero array in order to store the time for k-th arrival



for i in range(100000): #iterating the experiment 100000 times
    
    time_kth = 0 #initialize the value of time of k-th arrival
    
    for j in range(k):  #we will compute the time till 'k' arrivals happen
        
        
        time_kth += np.random.exponential(alpha) #the inter-arrival time in poisson process is exponentially distributed
    
    sample[i] = time_kth #appending the value of time taken along with its index i ( the ith iteration of experiment)


    

#Plotting the histogram 

plot.hist(sample, 50 ,color='brown',ec='white')

plot.title("TIME OF k-th ARRIVAL")

plot.ylabel("Frequency ------------->")

```
