```python
'''
Name: Karan Sutariya
Roll No: 2103322
CS230 Assignment[4]: Part III - Q1 Simulating Markov Chain and estimating the steady state probabilities of various states

'''


#Importing the following libraries to fabricate a markov chain

import numpy as np
import random 
import matplotlib.pyplot as plot 


p = q = 1/4 #probabilities 

iterations = 1000 #running the experiment for 1000 times for getting an accurate ans

initial_state = 0 #initial state 

markov_run = 100 #steps in MARKOV CHAIN

estimate = {} #creating a dictionary to store the state along with its count in 1000


for i in range(iterations): #iterating the experiment for 1000 times
    
    present_state = initial_state #initialization
    
    for x in range(markov_run): #running the markov chain for over 100 steps
        
        prob = random.uniform(0,1) #throws a random number in between 0 and 1
        
        if ( prob < p): 
            if (present_state != 4): present_state += 1  #moving to the next step
        
        elif ( prob < (p+q) ): 
            if (present_state != 0): present_state -= 1 #moving to the back step
                
    
    
    #inserting the states and their counts 
    if present_state in estimate: estimate[present_state] += 1
    
    else: estimate[present_state] = 1
        
 
 #for plotting the probabilities
for k in estimate.keys(): estimate[k] /= 1000
    
#Plotting the distribution
    
plot.bar(estimate.keys(), estimate.values(), color = 'pink', ec='black')

plot.xlabel("State  ------------->")

plot.ylabel("Steady State Probability  ------------->")

plot.title("ESTIMATION OF STEADY STATE PROBABILITY")

print("\nThe following steady state probabilities turn out to be a good approximation to the ones that are computed theoretically, i.e., 0.2. ")       
                
            
        
        
    
    

```
