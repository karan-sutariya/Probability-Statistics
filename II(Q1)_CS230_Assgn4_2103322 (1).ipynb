```python
'''
Name: Karan Sutariya
Roll No: 2103322
CS230 Assignment[4]: Part II - Q1 Simulating Poisson process and estimating the PMF of the combined queues via simulation

'''


#Importing the following libraries to work with poisson processes and plotting the distribution


import numpy as np
from numpy.random import poisson
import matplotlib.pyplot as plot

#define the rates of both the poisson processes

lambda1 = 1 # per hour
lambda2 = 2 # per hour

PMF = {}

#using this library we can generate random samples as arrays and then work with the arrays

process1 = np.random.poisson(lambda1,100000) 

process2 = np.random.poisson(lambda2,100000)

P = process1 + process2  #merging both the poisson processes ,i.e., P


#Inserting the entries of P in PMF along with their counts(as how many times do they appear) to plot it

for val in P:
        
    if val in PMF:
        PMF[val] = PMF[val] + 1
    else:
        PMF[val] =  1
        
for i in PMF.keys():
    PMF[i] = (PMF[i])/100000   #dividing by the total number of iterations


                
#Plotting the distribution

plot.bar(PMF.keys(), PMF.values(), color = 'green', ec='black')

plot.xlabel("Number of arrivals of the people ------------->")

plot.ylabel("Probability  ------------->")

plot.title("EMPIRICAL PMF DISTRIBUTION OF THE COMBINED QUEUES")














```

```python
'''
Plotting the theoretical PMF of a merged Poisson Process (P) whose rate parameter is sum of that of the poisson processes 1 and 2

'''

rate = 3 #per hour

P_theoretical = np.random.poisson( rate, 100000)
PMF_theoretical = {}

for val in P_theoretical:
        
    if val in PMF_theoretical:
        PMF_theoretical[val] += 1
    else:
        PMF_theoretical[val] =  1
        
for i in PMF_theoretical.keys():
    PMF_theoretical[i] *= (1/100000)   #dividing by the total number of iterations



#Plotting the distribution

plot.bar(PMF_theoretical.keys(), PMF_theoretical.values(), color = 'green', ec='black')

plot.xlabel("Number of arrivals of the people ------------->")

plot.ylabel("Probability  ------------->")

plot.title("THEORETICAL PMF DISTRIBUTION OF THE COMBINED QUEUES")
```
